Analyse av fiksjonell by i USA unde zombieapokoplypse 

```python
from pylab import *

#Betingelser
M = 500                         #Startpopulasjon
dM = 0                          #Forandring populasjon

invandrere = 0.2
fødte = 0.1  
k = (invandrere + fødte)/M      #Rate for positiv forandring av populasjon
b = 3000                        #Maks antall mennesker som kan bo i byen pga landområde

#Tid
t = 0                           #Starttid

tid = 365                       #Tid i antall dager
tim = tid*24                    #Tid i timer
dt = tid/tim                    #Tidsteg

#Gjør at en får 1/24 som forskjell i tid per intervall. 
#Dette gjøres til 7*24*(1/24) er gått, så man får se 
#forskjell i populasjon hver time over 7 dager. 


#Liste
timer = []
Mennesker = []
dMennesker = []

for i in range(tim):
    t = t + dt
    dM = k*M*(1-M/b)
    M = M + dM
    
    timer.append(t)
    dMennesker.append(dM)
    Mennesker.append(M)

plot(timer, Mennesker)
title('Populasjon av Alexandria hver time over 365 dager')
xlabel('Timer')
ylabel('Populasjon')
grid()
legend('mennesker')
```
