```python
#7.1, 7.13 1.
def derivert(f, x, dx):
    fder = (f(x+dx)-f(x))/dx
    return fder

def f(x):
    return 2*x + 1

print(derivert(f, 1, 1E-8))

xakse = []
yakse = []

tall = 0

for i in range(2,6,1):
    tall = derivert(f,i,1E-8)
    xakse.append(i)
    yakse.append(tall)
    
plot(xakse, yakse)
show()
```

```python
#7.1, 3.

def derivert(f, x, dx):
    fder = (f(x+dx)-f(x))/dx
    return fder

n = 1E-8
a = 1E+10
e = (1+1/a)**a

def f(x):
    global e
    return e**x

print(derivert(f, 1, n))

xakse = []
yakse = []

tall = 0

for i in range(2,6,1):
    tall = derivert(f,i,1E-8)
    xakse.append(i)
    yakse.append(tall)
    
plot(xakse, yakse)
show()
```
