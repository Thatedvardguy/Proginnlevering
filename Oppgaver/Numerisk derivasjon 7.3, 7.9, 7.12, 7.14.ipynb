```python
#7.3
from pylab import *
n = 1E+10
e = (1+1/n)**n

def N(t):
    return 0.2*(e**t) + 90

def derivert(f, x, dx):
    fder = (f(x+dx)-f(x))/dx
    return fder

xakse = []
yakse = []
derxakse = []
deryakse = []
tall1 = 0
tall2 = 0
for i in range (0, 10):
    tall1 = N(i)
    xakse.append(i)
    yakse.append(tall1)
    tall2 = derivert(N,i,1E-8)
    derxakse.append(i)    
    deryakse.append(tall2)
plot(xakse, yakse)
plot(derxakse, deryakse)
title("Plantepopulasjon")
legend(["Veksten", "Ending i veksten"])
show()
```

```python
from pylab import *

def derivert(f, x, dx):
    fder = (f(x+dx)-f(x))/dx
    return fder

def T(t):
    return 70*e**(-0.065*t)

tall = linspace(0,60,1000)
l = 1E-8

tempder = derivert(T,tall, l)
temp = T(tall)

plot(tall, tempder)
plot(tall, temp)
navn = ["Endring tempratur", "temperatur"]
legend(navn)
show()

print(f"Temperaturen endrer seg med {round(tempder[42])} når den når element 42.")
```

```python
from pylab import *
data = loadtxt('posisjon.txt', skiprows=1)
tid = data[:,0]
posisjon = data[:,1]

fart = []

for i in range(len(tid)-1):
    dy = posisjon[i+1] - posisjon[i]
    dx = tid[i+1] - tid[i]
    speed = dy/dx
    fart.append(speed)

akselerasjon = []

for i in range(len(tid)-2):
    dv_x = fart[i+1] - fart[i]
    dv_y = tid[i+1] - tid[i]
    aksel = dv_x/dv_y
    akselerasjon.append(aksel)

plot(tid, posisjon, label = "Posisjon")
plot(tid[:-1], fart, label = "Fart")
plot(tid[:-2], akselerasjon, label = "Akselerasjon")
xlabel("Tid")
ylabel("Meter")
legend()
show()
```
